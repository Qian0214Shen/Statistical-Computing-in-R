Q2.1:
Simulate a random variable $X$ with the following distribution law: $p_1=0.2$, $p_2=0.15$, $p_3=0.25$, $p_4=0.4$

```r
set.seed(123)
n=100;X=numeric()
for(i in 1:100){
  u=runif(1)
  if (u<0.2)X[i]=1
  else if(u<0.35)X[i]=2
  else if(u<0.6)X[i]=3
  else X[i]=4
}
table(X)
```

Q2.4: Random variable $X\sim b(n,p)$with the following distribution law:
$$p_i=P(X=i)=\frac{n!}{i!(n-i)!}p^i(1-p)^{n-1},i=0,1,\cdots,n.$$

```r
set.seed(123)
bino=function(n,p){
  Y=rep(0,n)
  for(j in 1:n){
    u=runif(1)
    i=0;pr=(1-p)^n;F=pr
    while (u>=F){
      pr=p*(n-i)*pr/(1-p)*(i+1)
      F=F+pr
      i=i+1
    }
    Y[j]=i
  }
  Y
}
table(bino(100,0.2))
```

Q2.9:Random variable $X\sim N(0,1)$, then the density function of $Z=|X|$ is:
$$f(z)=\frac{2}{\sqrt{2\pi}}e^{-\frac{z^2}{2}},0\lt z \lt \infty$$
Method steps for generating random numbers of $X$ using acceptance-rejection sampling.

```r
set.seed(2933)
p=function(m,d){
  n=c()
  for (i in 1:100){
    y=0
    u1=0
    u2=0
    while (u2<=exp(-(y-1)^2/2)){
      u1=runif(1)
      y=-log(u1)
      u2=runif(1)
    }
    z=y
    u3=runif(1)
    if(u3<=0.5) n=c(n,z)
    else n=c(n,-z)
  }
  zt=d*n+m
  zt
}
p(6,5)
mean(p(6,5))
```

Q2.13:
The density function of random variable $X$ is $f(x)$, and
$$f(x)=af_1(x)+(1-a)f_2(x), a\in(0,1)$$
$$f_1(x)=\frac{1}{\sqrt{2\pi}\sigma_1}e^{-\frac{(x-\mu_1)^2}{2\sigma^2_1}},f_2(x)=\frac{1}{\sqrt{2\pi}\sigma_2}e^{-\frac{(x-\mu_2)^2}{2\sigma^2_2}}$$
Generate random numbers for $X$ using the composition approach

```r
set.seed(2933)
p=function(m,d){
  n=c()
  for (i in 1:100){
    y=0
    u1=0
    u2=0
    while (u2<=exp(-(y-1)^2/2)){
      u1=runif(1)
      y=-log(u1)
      u2=runif(1)
    }
    z=y
    u3=runif(1)
    if(u3<=0.5) n=c(n,z)
    else n=c(n,-z)
  }
  zt=d*n+m
  zt
}

x1=p(1,2)
mean(x1)
```
